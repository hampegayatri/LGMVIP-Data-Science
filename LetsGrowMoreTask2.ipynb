# Image to Pencil Sketch with Python:

Importing Required libraries

```python
import cv2
import numpy as np
```

# Loading ImageÂ¶


```python
img=cv2.imread(r'C:\Users\MRUNAL\Downloads\shinchan.jpg',1000)
```

```python
img
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

Converting image into grey scale

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gray
```

```python
plt.imshow(img_gray)
```

# Inverting the image

```python
img_invert = cv2.bitwise_not(img_gray)
img_invert
```

```python
plt.imshow(img_invert)
```

smoothing the image

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
img_smoothing
```

```python
plt.imshow(img_smoothing)
```

```python
def dodgeV2(x, y):
    return cv2.divide(x, 255 - y, scale=256.0)
```

Final Sketch


```python
final_img = dodgeV2(img_gray, img_smoothing)
plt.imshow(final_img)
```

```python
cv2.imshow("Sketch", final_img)
cv2.waitKey(0)
```
